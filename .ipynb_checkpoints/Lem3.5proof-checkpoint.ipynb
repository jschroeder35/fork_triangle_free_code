```python
# Set up starting graph
v_0 = vert([1, 4, 5, 6], [], 0)  # v_1
v_1 = vert([2, 0], [], 1)  # v_2
v_2 = vert([3, 1], [], 2)  # v_3
v_3 = vert([4, 2], [], 3)  # v_4
v_4 = vert([0, 3], [], 4)  # v_5
v_5 = vert([0, 7], [], 5)  # a_1
v_6 = vert([0, 2, 7], [], 6)  # b_1
v_7 = vert([5, 6], [0, 1, 2, 3, 4], 7)  # u_1
test_graph = graph([v_0, v_1, v_2, v_3, v_4, v_5, v_6, v_7])
test_graph.update_graph(lem3_4_list)
test_graph.show()
```

```python
# Shows contradiction if c_5=c_2
test_graph2=copy.deepcopy(test_graph)
test_graph2.vertex_list[5].edges.append(3)
new_vert=vert([1, 4], [], 8)  # c_2/c_5
test_graph2.vertex_list.append(new_vert)
new_vert=vert([8], [0, 1, 2, 3, 4], 9)  # u_2
test_graph2.vertex_list.append(new_vert)
test_graph2.clean()
test_graph2.update_graph4(lem3_4_list, verbose=True, min_subgraph=C_5)
```
